### Question  6

Fit a linear regression model to predict the <code>'price'</code> using the feature <code>'sqft_living'</code> then calculate the R^2. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
#Enter Your Code, Execute and take the Screenshot
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
df = pd.read_csv('housing.csv')

# Select feature (sqft_living) and target (price)
X = df[['sqft_living']]
y = df['price']

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a linear regression model
model = LinearRegression()

# Fit the model using the training data
model.fit(X_train, y_train)

# Predict prices for the test data
y_pred = model.predict(X_test)

# Calculate R^2 score
r2 = r2_score(y_test, y_pred)

# Print the R^2 score
print(f'R^2 score: {r2}')
```

### Question 8

Use the list to create a pipeline object to predict the 'price', fit the object using the features in the list <code>features</code>, and calculate the R^2. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score

# Load your dataset into a DataFrame (replace 'housing.csv' with your actual data path)
df = pd.read_csv('housing.csv')

# Define features to be used for prediction
features = ['sqft_living', 'bedrooms', 'bathrooms']

# Select X and y
X = df[features]
y = df['price']

# Handle missing values (impute NaNs)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Create a pipeline with imputation, scaling, polynomial features, and linear regression
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Impute missing values
    ('scale', StandardScaler()),  # Standardize features
    ('polynomial', PolynomialFeatures(include_bias=False)),  # Generate polynomial features
    ('model', LinearRegression())  # Linear regression model
])

# Fit the pipeline to the data
pipeline.fit(X, y)

# Predict prices using the fitted pipeline
y_pred = pipeline.predict(X)

# Calculate R^2 score
r2 = r2_score(y, y_pred)

# Print the R^2 score
print(f'R^2 score: {r2}')
```

### Question 10

Perform a second order polynomial transform on both the training data and testing data. Create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2. You will need to submit it for the final project.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score


df = pd.read_csv('housing.csv')

# Select features and target
X = df[['sqft_living', 'bedrooms', 'bathrooms']]  # Adjust features as needed
y = df['price']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Perform polynomial transformation (degree=2) on both training and testing features
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Create Ridge regression object with alpha=0.1
ridge = Ridge(alpha=0.1)

# Fit the Ridge model using the polynomial transformed training data
ridge.fit(X_train_poly, y_train)

# Predict using the polynomial transformed test data
y_pred = ridge.predict(X_test_poly)

# Calculate R^2 score
r2 = r2_score(y_test, y_pred)

# Print the R^2 score
print(f'R^2 score: {r2}')
#Enter Your Code, Execute and take the Screenshot
```
